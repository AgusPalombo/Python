```python
import random
import os

# Configuración del tablero y variables iniciales
filas, cols = 20, 20
tablero = [[" " for _ in range(cols)] for _ in range(filas)]
cabeza_inicial = (filas // 2, cols // 2)
vibora = [cabeza_inicial]  # La vibora se representa como una lista de coordenadas
direccion = (0, 1)  # Dirección inicial: hacia la derecha

# Colocar comida en una celda vacía
espacios_vacios = [(r, c) for r in range(filas) for c in range(cols) if (r, c) not in vibora]
comida = random.choice(espacios_vacios) if espacios_vacios else None

while True:
    # Actualiza el tablero: lo limpia y coloca vibora y comida
    for r in range(filas):
        for c in range(cols):
            tablero[r][c] = " "
    tablero[vibora[0][0]][vibora[0][1]] = "O"  # Cabeza de la vibora
    for seg in vibora[1:]:
        tablero[seg[0]][seg[1]] = "o"
    if comida is not None:
        tablero[comida[0]][comida[1]] = "F"
    
    # Limpia la consola y muestra el tablero
    os.system('cls' if os.name=='nt' else 'clear')
    for r in range(filas):
        print(" ".join(tablero[r]))
    print("Movimiento (W=arriba, S=abajo, A=izquierda, D=derecha):")
    mov = input().upper()
    if mov == "W":
        nueva_dir = (-1, 0)
    elif mov == "S":
        nueva_dir = (1, 0)
    elif mov == "A":
        nueva_dir = (0, -1)
    elif mov == "D":
        nueva_dir = (0, 1)
    else:
        nueva_dir = direccion
    # Evita moverse en dirección opuesta
    if (nueva_dir[0] == -direccion[0] and nueva_dir[1] == -direccion[1]):
        nueva_dir = direccion
    else:
        direccion = nueva_dir
    
    # Calcula la nueva posición de la cabeza
    cabeza = vibora[0]
    nueva_cabeza = (cabeza[0] + direccion[0], cabeza[1] + direccion[1])
    
    # Verifica colisiones con bordes o consigo misma
    if (nueva_cabeza[0] < 0 or nueva_cabeza[0] >= filas or 
        nueva_cabeza[1] < 0 or nueva_cabeza[1] >= cols or 
        nueva_cabeza in vibora):
        os.system('cls' if os.name=='nt' else 'clear')
        for r in range(filas):
            print(" ".join(tablero[r]))
        print("¡Game Over!")
        break
    
    vibora.insert(0, nueva_cabeza)
    
    # Si la vibora come la comida, se coloca nueva comida; de lo contrario, se elimina la cola
    if comida is not None and nueva_cabeza == comida:
        espacios_vacios = [(r, c) for r in range(filas) for c in range(cols) if (r, c) not in vibora]
        if espacios_vacios:
            comida = random.choice(espacios_vacios)
        else:
            os.system('cls' if os.name=='nt' else 'clear')
            for r in range(filas):
                print(" ".join(tablero[r]))
            print("¡Ganaste! No hay más espacio para comida.")
            break
    else:
        vibora.pop()
```
